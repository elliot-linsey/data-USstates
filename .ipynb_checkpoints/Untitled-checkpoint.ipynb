```python
import pandas as pd
```

```python
areas = pd.read_csv('state-areas.csv')
areas = areas[:-1]
areas
```

```python
abbrev = pd.read_csv('state-abbrevs.csv')
```

```python
df = pd.merge(areas,abbrev,on='state')
```

```python
df['state'] = df['abbreviation']
df = df.drop(columns='abbreviation', axis=1)
```

```python
df2 = pd.read_csv('DRMS-14-MFOI-00004_2.csv')
df2.rename(columns={' TOTAL_COST .1':'total_cost1'},inplace=True)
df2.dropna(inplace=True)
df2
```

```python
df2.columns

#df2
```

```python
df2.total_cost1.value_counts()
```

```python
df2['total_cost1'] = df2['total_cost1'].str.replace(',', '').str.replace('$', '').str.replace(' ', '').str.replace('(', '').str.replace(')', '')
```

```python
df2[df2['total_cost1']==' -   ']
```

```python
df2['total_cost1'] = df2['total_cost1'].apply(lambda x: 0.00 if x == '-' else x)
df2['total_cost1'] = df2['total_cost1'].astype('float')
```

```python
group = df2.groupby('STATE')['total_cost1'].sum().reset_index()
group
```

```python
full = pd.merge(df, group, left_on='state', right_on='STATE')
full.drop(columns='STATE',axis=1,inplace=True)
full
```

```python
#full.to_excel('Areas Grouped.xlsx')
```

```python
pop = pd.read_csv('state-population.csv')
pop
```

```python
pop = pop[(pop['state/region'] != 'USA') & (pop['ages'] == 'total')]
pop
```

```python
group2 = pop.groupby('state/region')['population'].mean().reset_index()
group2
```

```python
full = pd.merge(full, group2, left_on='state', right_on='state/region')
full.drop(columns='state/region',axis=1,inplace=True)
full
```

```python
#full.to_excel('Areas Grouped.xlsx')
```

```python
(full.total_cost1/full.population).mean()
```
